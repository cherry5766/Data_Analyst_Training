```python
# cricket_analytics.py

def calculate_tax(amount, tax):
  tax = amount * tax
  return tax
def build_team_string(teams):
  team_string = ''
  for name in teams:
    name=name.get('team')
    team_string+=str(name)+" "
  return team_string
def get_highest_score(teams):
    top_score=[]
    for record in teams:
      record=record.get('highest_score')
      if record is not None:
          top_score.append(record)
    max_score=max(top_score)
    return max_score



if __name__ == "__main__":
    teams = [
        {"team": "MI", "city": "Mumbai", "highest_score": 220},
        {"team": "CSK", "city": "Chennai", "highest_score": 230},
        {"team": "RCB", "city": "Bangalore", "highest_score": 240},
    ]
    team_string = build_team_string(teams)
    top_score = get_highest_score(teams)
    tax = calculate_tax(500000, 0.18)
    print("Teams:", team_string)
    print("Highest score:", top_score)
    print("Tax on prize:", tax)
```
